```python
# for currency formatting
import locale
locale.setlocale( locale.LC_ALL, '' )

# for dataframe
import pandas as pd

# for colouring.
from termcolor import colored

# investment = money invested at first day.
# rate = percent of profit it will return in a day. for eg:- for 100 rs at 10% returns you'll have 110 rs at the day end.
# days_invested = no of days you kept your money invested. 

def Money(investment , profit_rate , trades):

    trade = 0
    total_fees_yet = 0
    profit_total = 0
    loss_total = 0
    tds = 0
    leverage = 20
    dollar = 79.71
    leveragefund = 0.004
    margin_coll = 0

    while trades > 0:
        trades = trades - 1
        trade = trade + 1
        initial_investment = investment # initial starting amount before anything
        if investment > 50000 * dollar:
            leverage = 20
            leveragefund = 0.005
            if investment > 250000 * dollar:
                leverage = 20
                leveragefund = 0.01
                if investment > 1000000 * dollar:
                    leverage = 20
                    leveragefund = 0.025
                    if investment > 10000000 * dollar:
                        leverage = 10
                        leveragefund = 0.05
                        if investment > 20000000 * dollar:
                            leverage = 5
                            leveragefund = 0.1
                            if investment > 50000000 * dollar:
                                leverage = 4
                                leveragefund = 0.125
                                if investment > 100000000 * dollar:
                                    leverage = 3
                                    leveragefund = 0.15
                                    if investment > 200000000 * dollar:
                                        leverage = 2
                                        leveragefund = 0.25
                                        if investment > 300000000 * dollar:
                                            leverage = 1
                                            leveragefund = 0.5

#         tds1 = investment * 0.01
#         investment = investment - tds1 # TDS cut
        
        margin_cut = investment * leverage * leveragefund
        margin_coll += margin_cut
        investment = (investment - margin_cut) * leverage
        margined_investment = investment
        
#         tds1 = margin_cut
#         investment = investment - tds1 # TDS cut after leveraged
        
        
        fees1 = ((0.04 * investment)/100) # fees charged for 1st taker/maker
        
        
        investment = investment - fees1
        total_fees_yet += fees1
        
        
        loss_rate = 0 #profit_rate/1 # 1.65 seems good Reward-to-Risk ratio as money does not decreases
        if trade % 100 == 0 and trade != 0:
            investment = investment - ((loss_rate/100) * investment) # stoploss after 1st fees charge
        else:
            investment = investment + ((profit_rate/100) * investment) # profit after 1st fees charge
        
        fees2 = ((0.04 * investment)/100) # fees charged for 2nd taker/maker
        investment = investment - fees2
        
        
        if trade % 100 == 0 and trade != 0:
            loss = investment - margined_investment
            loss_total -= loss
        else:
            profit = investment - margined_investment # total profit after 2nd fees charge
            profit_total += profit
        total_fees_yet += fees2

        
        profit_cent = ((profit * 100) / initial_investment)
        total_fees = fees1 + fees2
        
#         tds2 = investment * 0.01
#         investment = investment -  tds2 # TDS cut before removing leverage
       
        
        investment = investment - margined_investment + initial_investment
        
#         tds2 = investment * 0.01
#         investment = investment -  tds2 # TDS cut
        
#         tds += tds1 + tds2
        
        total_end_money = investment # last ending amount after everything
        per_trade_fees = (total_fees * 100)/initial_investment
        calculate_data = print("Trade -" ,  trade ,"#",colored("Start :","yellow","on_grey",attrs=['bold']), \
                               
        locale.currency(round(initial_investment,2), grouping=True) ,"|","$",round(initial_investment/dollar,3) ,"|", \
        
        colored(" End :","grey","on_yellow", attrs=['bold']) ,locale.currency(round(total_end_money,2), grouping=True) , \
                               
        "|", "$",round(total_end_money/dollar,3) ,"|", colored(" Profit :","red","on_white",attrs=['bold']), \
                               
        locale.currency(round(profit,2), grouping=True) ,"|" ,"$",round(profit/dollar,3), "|" ,round(profit_cent,2),"%", \
                               
        " | ","\n\n\tFees1 :",locale.currency(round(fees1,2), grouping=True) ," | ","Fees2 :", \
                               
        locale.currency(round(fees2,2), grouping=True), " | ","Total Fees :", locale.currency(round(total_fees,2), \
                                                                                              
        grouping=True)," | ","\n\n\tTotal Fees Yet :", locale.currency(round(total_fees_yet,2), grouping=True)," | ", \
                               
        round(per_trade_fees,3),"%"," | ",round(total_fees_yet*100/total_end_money,4), "%"," | ", "\n\n\tTotal Profit :", \
                               
        locale.currency(round(profit_total,2), grouping=True)," | ", "Total Loss :", locale.currency(round(loss_total,2), \
                                                                                                     
        grouping=True),
#                                " | ", "\n\n\tTOTAL TDS :",locale.currency(round(tds,2), grouping=True), " | ", \
                               
#         "TDS IN :",locale.currency(round(tds1,2), grouping=True), " | ", " TDS OUT :",locale.currency(round(tds2,2), \
                                                                                                      
#         grouping=True), " | "
                              )

        #         calculate_data = print(f"Trade -   {trade} # {colored(Start :,yellow,on_grey",attrs=['bold'])} ₹ {round(initial_investment,2)} | $ {round(initial_investment/75,3)} | {colored(" End :","grey","on_yellow",attrs=['bold'])} ₹ {round(total_end_money,2)} | $ {round(total_end_money/75,3)} | {colored(" Profit :","red","on_white",attrs=['bold'])} ₹ {round(profit,2)} |  $ {round(profit/75,3)} | {round(profit_cent,2)} % | \n\n\tFees1 : ₹ {round(-fees1,2)} | Fees2 : ₹ {round(-fees2,2)} | Total Fees : {round(-total_fees,2)} | Total Fees Yet : {round(total_fees_yet,2)} | {round(per_trade_fees,3)} % | {round(total_fees_yet*100/total_end_money,4)} %")
        print("\n-----------------------------------------------------------------------------------------------------------------\n")
```
